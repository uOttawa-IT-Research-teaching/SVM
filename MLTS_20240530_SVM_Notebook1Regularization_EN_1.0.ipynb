## About the synthetic dataset

Synthetic data is artificially generated, typically created using algorythms, rather than collected as representations of real-world phenomena. Synthetic datasets and randomly generated datasets used commonly for simulations and training machine learning models. Many times, that data only exists in memory for the duration of the program.

The make_blobs dataset is a synthetic dataset generator provided by scikit-learn, a popular machine learning library in Python. This dataset is particularly useful for learning and visualizing clustering and classification algorithms, such as Support Vector Machines (SVMs).

When you import make_blobs from sklearn.datasets, you get a function that generates randomly generated data points in the form of Gaussian blobs or clusters. Here's an overview of the make_blobs dataset:
1. **Dimensions**: The make_blobs function generates data points in a multi-dimensional space. By default, it creates a 2D dataset, but you can specify the desired number of features (n_features) to generate higher-dimensional data.
2. **Clusters**: The function creates a specified number of clusters or blobs (n_samples). Each cluster represents a different class or category in the dataset.
3. **Separation**: The clusters are separated from each other by a configurable amount, controlled by the cluster_std parameter. A lower cluster_std value results in well-separated clusters, while a higher value creates overlapping clusters, making the classification task more challenging.
4. **Cluster Centers**: The centers of the clusters are randomly generated within the bounding box specified by the center_box parameter.
5. **Randomness**: The make_blobs function generates random data points around the cluster centers, following a Gaussian distribution with the specified standard deviation (cluster_std).
6. **Class Labels**: Each data point is assigned a class label corresponding to the cluster it belongs to, making it a labeled dataset suitable for supervised learning tasks like classification.

### Linearly separable data​

In the Naive Bayes Classifier Tutorial, we saw an example of generative classification, where the ML model makes predictions to classify a specific example based on the probability of how features are distributed in the dataset. SVMs work through **discriminative classification,** making predictions by drawing a line between classes. In descriminative models, we simply find a line or curve (in two dimensions) or manifold (in multiple dimensions) that divides the classes from each other. Descrimination is harsh in real-world contexts, so another way of thinking about this is as a process of differentiating or recognizing differences, without implying that one set of phenomena is any less valuable than another. 

As an example of this, consider the simple case of a classification task in which the two classes of points are well separated, and therefore clearly distinct on a graph. 

Ask yourself the following question: <b>What boundary line could be drawn to classify the blue and green dots as separate and distinct?</b> (see the following figure):
   

```python

X, y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter');
```

A **linear discriminative classifier** would attempt to draw a straight line separating the two sets of data, and thereby create a model for classification.
For two-dimensional data like the one shown here, this is a task we could do by hand.

But immediately we see a complication: there is more than one possible dividing line that can create a decision boundary line between the two classes of blue dots and green dots!

We can draw some of them as follows; the following figure shows the result:

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')
plt.plot([0.6], [2.1], 'x', color='red', markeredgewidth=2, markersize=10)

for m, b in [(1, 0.65), (0.5, 1.6), (-0.2, 2.9)]:
    plt.plot(xfit, m * xfit + b, '-k')

plt.xlim(-1, 3.5);
```

Each line drawn between these data points classify the blue and green dots as different from eachother. These are three *very* different separators which, nevertheless, perfectly differentiate between these samples.
Depending on which you choose, a new data point (e.g., the one marked by the "X" in this plot) can be assigned a different label !
As with many real-world experiences of discrimination, the deceptively simple task of "drawing a line between classes" is actually complex and frought with unexpected outcomes, and we need to think more deeply about the concept of difference.

## SVM: Maximizing the Margin

Support vector machines offer one way to improve on this.
The intuition is this: rather than simply drawing a zero-width line between the classes, we can draw around each line a **margin** of some width, up to the nearest point. The best line would be the one that keeps the maximum distance from the closest members of both classes. This distance is called the "margin."

Here is an example of how this might look (see the following figure):

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')

for m, b, d in [(0.5, 1.6, 0.55)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',
                     color='blue', alpha=0.5)
for m, b, d in [(1, 0.65, 0.33), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',
                     color='orange', alpha=0.5)
plt.xlim(-1, 3.5);
```

The line that maximizes this margin is the one we will choose as the optimal model. In the diagram above, it's the blue one.

### Fitting a Support Vector Machine

Fitting data, or curve fitting, is the process of determining the model that best fits the curve of your series of data points, as you also saw in the Linear Regression Tutorial, the Naive Bayes Tutorial, and Decision Tree Tutorial. It is simply the process of training the model on the training dataset. 

#### Linearly separable data
Let's see the result of an actual fit to this data: we will use Scikit-Learn's support vector classifier (`SVC`) to train an SVM model on this data.
For the time being, we will use a **linear kernel** and set the **``C`` parameter** to a very large number (we'll discuss the meaning of these in more depth momentarily):

```python
from sklearn.svm import SVC # "Support vector classifier"
model = SVC(kernel='linear', C=1E10)
model.fit(X, y)
```

To better visualize what's happening here, let's create a quick convenience function that will plot SVM decision boundaries for us (see the following figure):

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, edgecolors='red',
                   facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')
plot_svc_decision_function(model);
```

This is the dividing line that maximizes the margin between the two sets of points.
Notice that a few of the training points just touch the margin: they are **circled in red** in the figure above.
These points are the pivotal elements of this fit; they are known as the `support vectors`, and give the algorithm its name.
In Scikit-Learn, the identities of these points are stored in the `support_vectors_` attribute of the classifier:

##### What do the support vectors represent here ?

Imagine you have two teams, Team A and Team B, playing a game in a field. The goal is to find a line that separates the players of Team A from the players of Team B. This line is called the "decision boundary."

Now, when you try to draw this line, you want to make sure that it is as far away from the players of both teams as possible. The players who are closest to the line from both teams are called the "support players."

The model.support_vectors_ is a list that contains the positions (coordinates) of these support players from both teams. These are the players who are closest to the decision boundary line and help define how far the line should be from each team.

The goal is to find a line (the decision boundary) that keeps the maximum distance from the support players of both teams. This distance is called the "margin." The larger the margin, the better the separation between the two teams, and the more confident you can be in deciding which team a new player belongs to.

When the computer is trying to find this decision boundary line, it looks at the positions of the support players from model.support_vectors_ and uses them to calculate the line that maximizes the margin between the two teams.

The support players in model.support_vectors_ are the most important players because they define the decision boundary line and the margin. The computer doesn't need to worry about the players who are far away from the line because they don't affect the decision boundary or the margin.

By using the positions of the support players, the computer can find the best line that separates the two teams with the maximum margin, making it easier to correctly identify which team a new player belongs to when they join the game.

In Python language, model.support_vectors_ is a NumPy array where each row represents a support vector. The number of rows in the array corresponds to the number of support vectors identified by the SVM model during training.

For example below, if you have a binary classification problem with two features (let's call them x and y), and the SVM model identifies 10 support vectors, then model.support_vectors_ will be a NumPy array with shape (10, 2), where each row contains the x and y values of a support vector


```python
# position in the form of an array
model.support_vectors_
```

A key to this classifier's success is that for the fit, only the positions of the support vectors matter; any points further from the margin that are on the correct side do not modify the fit.
Technically, this is because these points do not contribute to the function used to fit the model, so their position and number do not matter as long as they do not cross the margin.

We can see this, for example, if we plot the model learned from the first 60 points and first 120 points of this dataset (see the following figure):

```python
def plot_svm(N=10, ax=None):
    X, y = make_blobs(n_samples=200, centers=2,
                      random_state=0, cluster_std=0.60)
    X = X[:N]
    y = y[:N]
    model = SVC(kernel='linear', C=1E10)
    model.fit(X, y)
    
    ax = ax or plt.gca()
    ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    plot_svc_decision_function(model, ax)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)
for axi, N in zip(ax, [60, 120]):
    plot_svm(N, axi)
    axi.set_title('N = {0}'.format(N))
```

In the left panel, we see the model and the support vectors for 60 training points.
In the right panel, we have doubled the number of training points, but the model has not changed: the three support vectors in the left panel are the same as the support vectors in the right panel.
**This insensitivity to the exact behavior of distant points is one of the strengths of the SVM model.**

### Tuning the SVM: Softening Margins

#### Non-linearly separable data
Our discussion thus far has centered around "clean" datasets with very little noise, in which a perfect, linear decision boundary can be drawn.

But what if your data has some amount of overlap and is **non-linearly separable data?**

For example, you may have data like this (see the following figure):

```python
X, y = make_blobs(n_samples=100, centers=2,
                  random_state=0, cluster_std=1.2)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter');
```

Several blue dots fall in the category of the green dots and vice versa. Data separation is not perfect and will need to tolerate some outliers in the classification. Instead of a straight boundary line, we could draw a zigzag line on the graph to separate the classes, but zigzags make a model difficult to train. 

#### 'C' parameter

To handle this case, the SVM implementation has a parameter that can compensate for error as it allows some margin violations (misclassifications) that is, it allows some of the points to creep into the margin, if that produces a better fit between the model and the data.

The tolerance of those violations is controlled by a tuning parameter, most often known as `C`.
For a very large `C`, the margin violations tolerance is stricter, and points cannot lie in it.
For a smaller `C`, the margin violations tolerance is less strict and can encompass some points.

The plot shown in the following figure gives a visual picture of how a changing `C` affects the final fit via the softening of the margin:

```python
X, y = make_blobs(n_samples=100, centers=2,
                  random_state=0, cluster_std=0.8)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for axi, C in zip(ax, [10.0, 0.1]):
    model = SVC(kernel='linear', C=C).fit(X, y)
    axi.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')
    plot_svc_decision_function(model, axi)
    axi.scatter(model.support_vectors_[:, 0],
                model.support_vectors_[:, 1],
                s=300, lw=1, facecolors='none');
    axi.set_title('C = {0:.1f}'.format(C), size=14)
```

#### Cross-validation

The optimal value of `C` will depend on your dataset, and you should tune this parameter using **cross-validation** or a similar procedure.

Cross-validation helps prevent overfitting (where the model performs well on the training data but poorly on new data) and ensures that the chosen value of `C` generalizes well to new, unseen data.

To find the best value of the `C`parameter, you can use a technique called cross-validation. Here's how it works:
 
1.  Divide the group of students into smaller groups (folds).
    
2.  For each possible value of `C` (e.g., 0.1, 0.5, 1.0, 10.0, 100.0): 
 * a. Use one of the folds as a "test" group and the remaining folds as a "training" group. 
 * b. Train the SVM model on the "training" group using the current value of `C`. 
 * c. Test the trained model on the "test" group and calculate how many students were misclassified. 
 * d. Repeat steps a-c for each fold, using a different fold as the "test" group each time. 
 * e. Calculate the average number of misclassified students across all folds for the current value of `C`.
    
3.  Choose the value of `C` that results in the lowest average number of misclassified students.
    
 
The idea behind cross-validation is to simulate the real-world scenario where you have to make predictions on new, unseen data (the "test" folds). By trying different values of `C` and evaluating the performance on the "test" folds, you can find the value of `C` that gives the best balance between separating the teams well and minimizing misclassifications.
 
The value of `C` that achieves this balance is considered the optimal value, and it can be used to train the final SVM model on the entire group of students.
